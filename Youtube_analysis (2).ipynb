```python
#get these data in one table 'views','likes','dislikes','comment_count','Genre'          predicted likes is output/dependent and remaining were input/independent variable

filtered=df2[['views','likes','dislikes','comment_count','Genre']]
filtered
```

```python
#We have two methods. 1.Label encoding and 2. one -hot encoding
#1.Label encoding limitation is prioritizing the value.
#2.one hot encoding --dummy variable trap
```

```python
#Do one Hot encoding for Genre.


f1=pd.get_dummies(filtered['Genre'])
f1
```

```python
#Drop one any column for dummy variable trap in one hot encoding.

f1.drop("Travel & Events",axis=1,inplace=True)
```

```python
f1
```

```python
#Concatenate the filtred and f1 (after dropping the coulmn).

fi=pd.concat([filtered,f1],axis=1)
fi
```

```python
fi.drop("Genre",axis=1,inplace=True)
fi
```

```python
fi
```

```python
#Writing back to dataframe to csv(use -to_csv)

fi.to_csv("filteredutube.csv",index=False)
```

```python
#x is input data/independent variable
#y is output data/dependent variable
```

```python
x=fi.drop("likes",axis=1)
```

```python
y=fi["likes"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Creating a object for linearRegression

lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test.head())
```

```python
##without np .array it gives row wise;

np.array(y_test.head())
```

```python
lr.score(x_test,y_test)
```
