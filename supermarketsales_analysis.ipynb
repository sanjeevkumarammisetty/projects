```python
import ipywidgets as widgets
```

```python
#Based upon the city

df["City"].unique()
```

```python
c=widgets.Dropdown(
    options=df['City'].unique(),
    value='Boston',
    description='City:',
    disabled=False,
)

def city(cityname):
  temp=df[df["City"]==cityname]
  plt.title(cityname)
  sns.barplot(y=df["Product"],x=df["Total Sales"],estimator=sum)

```

```python
widgets.interact(city,cityname=c)
```
